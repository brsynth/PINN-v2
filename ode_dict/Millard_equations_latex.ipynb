## ODE system

Differential equations in Millard's model. Orange terms are required only to simulate chemostat experiments. They can be replaced by 0 when unused.

$$
\begin{equation}
\begin{split}

\frac{dGLC}{dt} &=
-\nu_{glycolysis} \cdot X \cdot \frac{V_{cell}}{V_{env}} \textcolor{orange}{+ \nu_{feed} - D \cdot GLC} \\[3ex]

\frac{dACE_{env}}{dt} &=
\nu_{acetate\_exchange} \cdot X \cdot \frac{V_{cell}}{V_{env}} \textcolor{orange}{- D \cdot ACE_{env}} \\[3ex]

\frac{dX}{dt} &= 
X \cdot \nu_{growth} \textcolor{orange}{- D \cdot X} \\[3ex]

\frac{dACCOA}{dt} &=
1.4 \cdot \nu_{glycolysis} - \nu_{TCA_{cycle}} - \nu_{Pta} \\[3ex]

\frac{dACP}{dt} &=
\nu_{Pta} - \nu_{AckA} \\[3ex]

\frac{dACE_{cell}}{dt} &=
\nu_AckA - \nu_{acetate\_exchange}

\end{split}
\end{equation}
$$

Reaction rates have the following rate laws. Orange terms can be replaced with 1 when ignoring the inhibition of the glycolytic and/or TCA cycle pathways by acetate.
$$
\begin{equation}
\begin{split}

\nu_{glycolysis} &=
\frac{V_{max_{glycolysis}} \cdot GLC}{GLC + Km\_GLC} \textcolor{orange}{\cdot \frac{1}{1 + \frac{ACE_{env}}{Ki\_ACE}}} \\[5ex]

\nu_{TCA_{cycle}} &=
\frac{V_{max_{TCA_{cycle}}} \cdot ACCOA}{ACCOA + Km\_ACCOA} \textcolor{orange}{\cdot \frac{1}{1 + \frac{ACE_{env}}{Ki\_ACE}}} \\[5ex]

\nu_{AckA} &=
\frac{\frac{V_{max_{AckA}} \cdot \left( ACP \cdot ADP - \frac{ACE_{cell} \cdot ATP}{Keq} \right)}{Km\_ACP \cdot Km\_ADP}}{\left(1 + \frac{ACP}{Km\_ACP} + \frac{ACE_{cell}}{Km\_ACE}\right) \cdot \left(1 + \frac{ADP}{Km\_ADP} + \frac{ATP}{Km\_ATP}\right)} \\[5ex]

\nu_{Pta} &=
\frac{\frac{V_{max_{Pta}} \cdot \left( ACCOA \cdot P - \frac{ACP \cdot COA}{Keq} \right)}{Km\_ACCOA \cdot Km\_P}}{1 + \frac{ACCOA}{Km\_ACCOA} + \frac{P}{Ki\_P} + \frac{AcP}{Ki\_AcP} + \frac{COA}{Km\_COA} + \frac{ACCOA \cdot P}{Km\_ACCOA \cdot Km\_P} + \frac{ACP \cdot COA}{Km\_ACP \cdot Km\_COA}} \\[5ex]

\nu_{acetate\_exchange} &=
\frac{\frac{V_{max_{acetate\_exchange}} \cdot \left( ACE_{cell} - \frac{ACE_{env}}{Keq} \right)}{Km\_ACE}}{1 + \frac{ACE_{cell}}{Km\_ACE} + \frac{ACE_{env}}{Km\_ACE}} \\[5ex]

\nu_{growth} &= \nu_{TCA_{cycle}} \cdot Y

\end{split}
\end{equation}
$$
